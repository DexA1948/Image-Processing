#Cross_Shaped_Kernel

```python
#3_x_3_Kernel
kernel3 = cv.getStructuringElement(cv.MORPH_CROSS,(5,5))
print(kernel3)
```

#Performing_Erosion_operation_on_chipimg

```python
#2_Iterations
eroded_chipimg = cv.erode(chipimg, kernel3, iterations = 2)
cv2_imshow(chipimg)
cv2_imshow(eroded_chipimg)
```

```python
#10_iterations
eroded_chipimg = cv.erode(chipimg, kernel3, iterations = 10)
cv2_imshow(eroded_chipimg)
```

#Performing_Dilation_operation_on_textimg

```python
#1_Iteration
dilated_textimg = cv.dilate(textimg, cv.getStructuringElement(cv.MORPH_CROSS,(5,5)), iterations = 1)
cv2_imshow(textimg)
cv2_imshow(dilated_textimg)
```

#Performing_Close_operation_in_pepperimg

```python
#Using_3_x_3_kernel
closing = cv.morphologyEx(pepimg, cv.MORPH_CLOSE, kernel3)
cv2_imshow(pepimg)
cv2_imshow(closing)
```

```python
#Using_5_x_5_kernel
closing = cv.morphologyEx(pepimg, cv.MORPH_CLOSE, kernel5)
cv2_imshow(closing)
```
