```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread("/content/drive/MyDrive/Me.jpg")
```

```python
image_gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
(thresh, image_bw) = cv2.threshold(image_gray, 127, 255, cv2.THRESH_BINARY)
```

```python
cv2_imshow(image)
```

```python
cv2_imshow(image_gray)
```

```python
cv2_imshow(image_bw)
```

```python
image_edge=cv2.Canny(image_gray,0,100)
```

```python
cv2_imshow(image_edge)
```
