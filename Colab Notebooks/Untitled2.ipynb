```python
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
pepimg = cv.imread("/content/drive/MyDrive/ipphotos/tilljune24/close.png", 0)
```

```python
pepimg
```

```python
cv2_imshow(pepimg)
```

```python
kernel3 = cv.getStructuringElement(cv.MORPH_CROSS, (5,5))
```

```python
kernel3
```

```python
eroded = cv.erode(pepimg, kernel3, iterations=1)
```

```python
cv2_imshow(eroded)
```

```python
dilated = cv.dilate(pepimg, kernel3, iterations=1)
```

```python
cv2_imshow(dilated)
```
